# Kullanıcıdan Celcius bilgisini alıp fahrenheita çeviren kodu yazın

```python
celc = float(input("lütfen hava sıcaklığını celcius cinsinden girin örn: 25.5"))

fahr = celc * 1.8 + 32

print("Hava sıcaklığı", celc, "derece", fahr, "fahrenheit eder")
```

```python
celc = float(input("lütfen hava sıcaklığını celcius cinsinden girin örn: 25.5"))

fahr = celc * 1.8 + 32

print("Hava sıcaklığı", celc, "derece", round(fahr,2), "fahrenheit eder")
```
